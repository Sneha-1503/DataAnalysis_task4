```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("products-100.csv")
df.head()
```

```python
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
```

```python
df.isnull().sum()
```

```python
category_count = df['Category'].value_counts()

print(category_count)
```

```python
category_count.plot(kind='bar')
plt.title("Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
brand_count = df['Brand'].value_counts()

brand_count.plot(kind='bar')
plt.title("Products by Brand")
plt.show()
```

```python
sns.countplot(data=df, x='Availability')
plt.title("Product Availability")
plt.show()
```

```python
sns.histplot(df['Price'], bins=15, kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
sns.histplot(df['Stock'], bins=15, kde=True)
plt.title("Stock Distribution")
plt.show()
```

```python
sns.scatterplot(data=df,
                x='Price',
                y='Stock')

plt.title("Price vs Stock")
plt.show()
```

```python
corr = df[['Price','Stock']].corr()

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
